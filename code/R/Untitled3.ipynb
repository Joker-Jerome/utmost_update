```r
# expression check 
library(data.table)
library(dplyr)
library(ggplot2)
```

```r
df_exp <- as.data.frame(fread("/ysm-gpfs/pi/zhao-data/zy92/GTEx_V8/GTEx_Analysis_2017-06-05_v8_RNASeQCv1.1.9_gene_tpm.gct"))
```

```r
df_exp
```

```r
interesting_exp <- df_exp %>%
    filter(Description %in% c("DROSHA", "DICER1", "INO80", "MYC")) 
```

```r
dim(interesting_exp)
```

```r
interesting_exp_sub <- interesting_exp[, -1]
```

```r
df <- t(interesting_exp_sub)
```

```r
head(df)
```

```r
df <- as.data.frame(df)
```

```r
colnames(df) <- c("DROSHA", "DICER1", "INO80", "MYC")
```

```r
df <- df[-1, ]
```

```r
df <- apply(df, 2, as.numeric)
```

```r
df <- as.data.frame(df)
```

```r
p<-ggplot(df, aes(x=DROSHA)) + 
  geom_histogram(color="black", fill="white")

p+ geom_vline(aes(xintercept=mean(DROSHA)),
            color="blue", linetype="dashed", size=1)
# Histogram with density plot
ggplot(df, aes(x=DROSHA)) + 
 geom_histogram(aes(y=..density..), colour="black", fill="white")+
 geom_density(alpha=.2, fill="#FF6666") 
```
